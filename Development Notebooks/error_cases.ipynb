```python
def get_molecular_formala(mol):
    h_ats = 0
    o_ats = 0
    n_ats = 0
    c_ats = 0
    s_ats = 0
    for atom in mol.GetAtoms():
        h_ats += atom.GetTotalNumHs()
        symbol = atom.GetSymbol()
        if symbol == "O":
            o_ats += 1
        elif symbol == "C":
            c_ats += 1
        elif symbol == "N":
            n_ats += 1
        elif symbol == "S":
            s_ats += 1
    output = f'C{c_ats}H{h_ats}'
    if n_ats: 
        output+= f'N{n_ats}'
    elif o_ats:
        output += f'O{o_ats}'
    elif s_ats:
        output += f'S{s_ats}'
    return output
```
